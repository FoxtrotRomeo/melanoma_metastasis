```python
import matplotlib
if 'init_done' in globals():
    matplotlib.use("pgf")
    matplotlib.rcParams.update({
        "pgf.texsystem": "pdflatex",
        'font.family': 'serif',
        'text.usetex': True,
        'pgf.rcfonts': False,
    })
import matplotlib.pyplot as plt
import numpy as np

from matplotlib.colors import ListedColormap
from typing import Iterable

init_done = True
```

```python
def plot_table(ax:plt.axes, labels:Iterable[str], array:np.array, threshold:float=0.05):
    n = len(labels)
    assert len(array.shape) == 2
    assert array.shape[0] == n
    assert array.shape[1] == n

    # Disable axes
    ax.set_axis_off()

    # Plot table
    ax.imshow(array <= threshold, cmap=ListedColormap(["white", "grey"]))
    ax.bar(np.arange(n+1)-.5,
        n, .015, -.5, color='black'
    )
    ax.barh(np.arange(n+1)-.5,
        n, .015, -.5, color='black'
    )

    # Print labels:
    for x,s in enumerate(labels):
        ax.text(
            x, -.7, s,
            fontweight='bold',
            ha="center", va="center"
        )
    for y,s in enumerate(labels):
        ax.text(
            -.7, y, s,
            fontweight='bold',
            ha="center", va="center",
            rotation=90
        )

    # Loop over data dimensions and create text annotations.
    for i in range(len(labels)):
        for j in range(len(labels)):
            ax.text(j, i, f"{array[i, j]:.4f}", ha="center", va="center", size=14)
```

```python
ax = plt.gca()

plot_table(
    ax,
    ['SVM', 'RF', 'DT'],
    np.array([
        [1.0000, 0.0071, 0.0010],
        [0.0070, 1.0000, 0.0184],
        [0.0010, 0.0184, 1.0000]
    ])
)

plt.show()
```

```python
titlefont = {'fontsize': 16, 'fontweight': 'bold'}

fig = plt.figure(figsize=[15,12])

ax = plt.subplot(2,3,1)
ax.set_title('a)', loc='left', fontdict=titlefont)
plot_table(
    ax,
    ['SVM', 'RF', 'DT'],
    np.array([
        [1.0000, 0.0071, 0.0010],
        [0.0070, 1.0000, 0.0184],
        [0.0010, 0.0184, 1.0000]
    ])
)

ax = plt.subplot(2,3,2)
ax.set_title('b)', loc='left', fontdict=titlefont)
plot_table(
    ax,
    ['LR', 'DT', 'NN\n2 layers'],
    np.array([
        [1.0000, 0.3191, 0.0010],
        [0.3191, 1.0000, 0.0010],
        [0.0010, 0.0010, 1.0000]
    ])
)

ax = plt.subplot(2,3,3)
ax.set_title('c)', loc='left', fontdict=titlefont)
plot_table(
    ax,
    ['KNN', 'Catch22', 'LSTM\n1 layer'],
    np.array([
        [1.0000, 0.0035, 0.0035],
        [0.0035, 1.0000, 0.0010],
        [0.0035, 0.0010, 1.0000]
    ])
)

ax = plt.subplot(2,3,4)
ax.set_title('d)', loc='left', fontdict=titlefont)
plot_table(
    ax,
    ['Transformer\n+ BiLSTM', 'Transformer\n+ LSTM', 'Transformer\n+ GRU'],
    np.array([
        [1.0000, 0.7260, 0.1790],
        [0.7260, 1.0000, 0.5471],
        [0.1790, 0.5471, 1.0000]
    ])
)

ax = plt.subplot(2,5,(9,10))
ax.set_title('e)', loc='left', fontdict=titlefont)
plot_table(
    ax,
    ['Time Series\nText', 'Static Data\nText', 'Static Data\nTime Series', 'All'],
    np.array([
        [1.0000, 0.0243, 0.0010, 0.0594],
        [0.0243, 1.0000, 0.1285, 0.9000],
        [0.0010, 0.1285, 1.0000, 0.0594],
        [0.0594, 0.9000, 0.0594, 1.0000]
    ])
)

plt.tight_layout()
plt.savefig('pictures/significance.pdf')
```
